if elif code

```python
choice = input('Enter your choice : +,-,*,/')
number1 = int(input('Enter number 1'))
number2 =int( input('Enter number 2'))
if choice == "+" :
    print(f"The sum of the numbers is: {number1+number2}")
elif choice == "-" :
    print(f"The difference of the numbers is: {number1-number2}")
elif choice == "*" :
    print(f"The product of the numbers is: {number1*number2}")
elif choice == "/" :
    print(f"The quotient of the numbers is: {number1/number2}")
else :
    print('Please enter a valid choice')
```

**Nested if else**

```python
choice = input("Enter a choice : Coffe or Tea")
if choice == "coffe" or "c" :
    preference = input('Enter your preference :you want milk ? yes or no? ')
    if preference == "yes" :
        print("You added the milk to coffe")
    elif preference == "no" :
        print("You didn't add the milk")
elif choice == "Tea or t" :
    preference = input("do you want sugar ? yes or no? ")
    if preference == "yes" :
        print("You added the sugar")
    elif preference == "no" :
        print("You didn't add the sugar")
else :
    print('Please enter a valid choice')
```
